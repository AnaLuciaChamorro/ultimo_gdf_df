```python
#ejercicio 1
#Subimos la hoja de cálculo
import pandas as pd
import os

datadis = pd.read_excel(os.path.join('data', 'reporte_estadist.xlsx'), dtype={'Ubigeo': object})
datadis.head()
```

```python
# mapa provincias

os.environ['USE_PYGEOS'] = '0'

import geopandas as gpd

datadismap=gpd.read_file(os.path.join('maps','DistritosMap.zip'))

datadismap.head()
```

```python
#poner todo en mayúscula y eliminar los espacios en blanco alrededor de los valores
capitalizeColumns=lambda x: x.str.upper().str.strip()
datadis[['Provincia','Distrito']]=datadis[['Provincia','Distrito']].apply(capitalizeColumns)
datadismap[['PROVINCIA','DISTRITO']]=datadismap[['PROVINCIA','DISTRITO']].apply(capitalizeColumns)
```

```python
import unidecode

# Definimos la función byePunctuation
byePunctuation = lambda x: unidecode.unidecode(str(x))

# Aplicar la función a las columnas 'Provincia' y 'Distrito', y eliminamos los símbolos
datadis[['Provincia','Distrito']] = datadis[['Provincia','Distrito']].applymap(byePunctuation)
datadismap[['PROVINCIA','DISTRITO']] = datadismap[['PROVINCIA','DISTRITO']].applymap(byePunctuation)
```

```python
#numero de filas de la data de mapa y data
len(datadis.Distrito),len(datadismap.DISTRITO)
```

```python
#vemos cuantos repetidos hay
datadis.Distrito.duplicated().sum(),datadismap.DISTRITO.duplicated().sum()
```

```python
#crear una nueva columna que une la provincia y el distrito
datadis['provDist']=["+".join(pd) for pd in zip (datadis.Provincia,datadis.Distrito)]
datadismap['provDist']=["+".join(pd) for pd in zip (datadismap.PROVINCIA,datadismap.DISTRITO)]
```

```python
#vemos cómo se ve la nueva columna
datadis['provDist'].head()
```

```python
#reemplazamos los guiones por espacios
datadis.provDist=datadis.provDist.str.replace("\-|\_|\s+"," ",regex=True)
datadismap.provDist=datadismap.provDist.str.replace("\-|\_|\s+"," ",regex=True)
```

```python
#identificamos que datos no están igual escritos en ambas datas
nomatch_df=set(datadis.provDist)- set(datadismap.provDist)
nomatch_gdf=set(datadismap.provDist)-set(datadis.provDist) 
```

```python
#cantidad de distritos que están escritos diferente
len(nomatch_df), len(nomatch_gdf)
```

```python
from thefuzz import process
#buscar el distrito que más se parece a los 8 anteriores, los numeros que salen al costado es el porcentaje que se parecen
[(dis,process.extractOne(dis,nomatch_gdf)) for dis in sorted(nomatch_df)]
#lista
```

```python
#creamos un diccionario con las coincidencias más cercanas
changesDis_df={dis:process.extractOne(dis,nomatch_gdf)[0] for dis in sorted(nomatch_df)}
```

```python
#unimos la data (reemplazamos porvDist pot changesDis)
datadis.provDist.replace(changesDis_df,inplace=True)
```

```python
#metemos el mapa a la tabla
datadismap=datadismap.merge(datadis, on='provDist')
```

```python
#toda las maneras de hacer intervalos
import mapclassify 
import numpy as np

np.random.seed(12345) # so we all get the same results!

#Trabajaremos con 5 intervalos
K=5
#clasificación con intervalos iguales
ei5 = mapclassify.EqualInterval(junin['Viv_agua_dia'], k=K)
#mismo ancho de intervalo basado en la desviación estándar
msd = mapclassify.StdMean(junin['Viv_agua_dia'])
#clasificación de intervalos por cuantiles
q5=mapclassify.Quantiles(junin['Viv_agua_dia'],k=K)

#usamos el método MaxiumBreaks, con la varianza mínima
mb5 = mapclassify.MaximumBreaks(junin['Viv_agua_dia'], k=K)
#la cantidad de intervalos se forman por el algoritmo, usa también los datos extremos
ht = mapclassify.HeadTailBreaks(junin['Viv_agua_dia'])
# based on similarity, optimizer
fj5 = mapclassify.FisherJenks(junin['Viv_agua_dia'], k=K)
# based on similarity, optimizer
jc5 = mapclassify.JenksCaspall(junin['Viv_agua_dia'], k=K)
# based on similarity, optimizer
mp5 = mapclassify.MaxP(junin['Viv_agua_dia'], k=K) 
```

```python
import numpy as np
from scipy.spatial import distance

# Datos de ejemplo
junin_area_verd_consv = junin['Area_verd_consv'].values
junin_area_verd_consv = junin_area_verd_consv.astype(float)  # Convertir a tipo float
w_knn = 4  # Número de vecinos

# Calcular la matriz de distancias
dist_matrix = distance.squareform(distance.pdist(junin_area_verd_consv.reshape(-1, 1)))

# Calcular la matriz de pesos basada en la distancia
w = np.zeros_like(dist_matrix)
w[np.argsort(dist_matrix, axis=1)[:, :w_knn+1]] = 1

# Calcular el índice de autocorrelación de Moran
n = len(junin_area_verd_consv)
x_bar = np.mean(junin_area_verd_consv)
S0 = np.sum(w)
S1 = np.sum(w * (junin_area_verd_consv - x_bar))
S2 = np.sum(w * (junin_area_verd_consv - x_bar)**2)

# Realizar permutaciones y calcular el valor p simulado
num_permutations = 999  # Número de permutaciones
count_bigger = 0  # Contador de permutaciones con S1 mayor o igual que S1_permuted

for _ in range(num_permutations):
    permuted_consv = np.random.permutation(junin_area_verd_consv)
    S1_permuted = np.sum(w * (permuted_consv - x_bar))
    if S1_permuted >= S1:
        count_bigger += 1

moran_p_sim_AVC = count_bigger / num_permutations
moran_I_AVC = (n / S0) * (S1 / S2)
moran_I_AVC,moran_p_sim_AVC
```

```python
import matplotlib.pyplot as plt

# Calcular el lag espacial estandarizado
spatial_lag_AVC_std = np.dot(w, junin_area_verd_consv) / S0

# Gráfico de dispersión espacial
plt.scatter(junin_area_verd_consv, spatial_lag_AVC_std)
plt.xlabel('AVC_std')
plt.ylabel('Spatial Lag AVC_std')
plt.title('Gráfico de Dispersión Espacial')
plt.gca().set_aspect('equal', adjustable='box')  # Aspecto igual en ambos ejes
plt.show()
```

```python
import numpy as np
from scipy.spatial import distance

# Datos de ejemplo
junin_area_verd_consv = junin['Area_verd_consv'].values
junin_area_verd_consv = junin_area_verd_consv.astype(float)  # Convertir a tipo float
w_knn = 6  # Número de vecinos (modifica este valor según sea necesario)

# Calcular la matriz de distancias
dist_matrix = distance.squareform(distance.pdist(junin_area_verd_consv.reshape(-1, 1)))

# Calcular la matriz de pesos basada en la distancia
w = np.zeros_like(dist_matrix)
w[np.argsort(dist_matrix, axis=1)[:, :w_knn+1]] = 1

# Verificar si la matriz de pesos es singular
if np.linalg.cond(w) < 1 / np.finfo(float).eps:
    print("La matriz de pesos es singular. Verifica tus datos.")

# Calcular el lag espacial estandarizado
n = len(junin_area_verd_consv)
x_bar = np.mean(junin_area_verd_consv)
S0 = np.sum(w)
S1 = np.sum(w * (junin_area_verd_consv - x_bar))
S2 = np.sum(w * (junin_area_verd_consv - x_bar)**2)
spatial_lag_AVC_std = np.dot(w, junin_area_verd_consv) / S0

# Gráfico de dispersión espacial
plt.scatter(junin_area_verd_consv, spatial_lag_AVC_std)
plt.xlabel('AVC_std')
plt.ylabel('Spatial Lag AVC_std')
plt.title('Gráfico de Dispersión Espacial')
plt.show()
```

```python
import numpy as np

# Calcular el rango intercuartil
Q1 = np.percentile(junin['Area_verd_consv'], 25)
Q3 = np.percentile(junin['Area_verd_consv'], 75)
IQR = Q3 - Q1

# Definir un umbral para identificar outliers
threshold = 1.5  # Puedes ajustar este valor según tus necesidades

# Encontrar los índices de los valores atípicos
outliers = np.where((junin['Area_verd_consv'] < Q1 - threshold * IQR) | (junin['Area_verd_consv'] > Q3 + threshold * IQR))[0]
```
